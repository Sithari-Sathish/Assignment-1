```python
##simple examples

x=np.arange(0,10)
y=np.arange(11,21)
print(x)
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'go--', linewidth=1, markersize=7)
plt.title('Line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
x = np.arange(0, 4 * np.pi , 0.1)
y= np.sin(x)
plt.title("sine waveform")

#plot the points using matlotlib
plt.plot(x,y,color='b')
plt.show()
```

```python
x= np.arange(0,10)
y=x*x
y
```

```python
plt.plot(x,y)
```

```python
plt.title('square plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
plt.show()
```

```python
#creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y, 'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```
